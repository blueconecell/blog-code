```python
import random
```

```python
def Intro():
  print('H A N G M A N')
```

```python
# 처음상태 0, 최후상태 6
man =['''
  +---+
      |
      |
      |
    ===
''','''
  +---+
  o   |
      |
      |
    ===
''','''
  +---+
  o   |
  |   |
      |
    ===
''','''
  +---+
  o   |
 /|   |
      |
    ===
''','''
  +---+
  o   |
 /|\  |
      |
    ===
''','''
  +---+
  o   |
 /|\  |
 /    |
    ===
''','''
  +---+
  o   |
 /|\  |
 / \  |
    ===
''']
```

```python
word = '''ant baboon badger bat bear beaver camel cat clam cobra cougar
       coyote crow deer dog donkey duck eagle ferret fox frog goat goose hawk
       lion lizard llama mole monkey moose mouse mule newt otter owl panda
       parrot pigeon python rabbit ram rat raven rhino salmon seal shark sheep
       skunk sloth snake spider stork swan tiger toad trout turkey turtle
       weasel whale wolf wombat zebra'''.split()
```

```python
def RnWord():     # 정답이 될 단어를 추출해주는 함수
  return random.choice(word)
```

```python
def Board(string):    # 현재 진행되는 board 초기상태를 만들어주는 함수
  arr = []
  for _ in string:
    arr.append('_')
  return arr
```

```python
def Find(string, char):   # 추측한 낱말이 몇번 인덱스에 있는지 찾아주는 함수
    for i, c in enumerate(string):
        if c == char:
            yield i
```

```python
def Correct(boardList, idxList, guess):   # 정답을 맞추면 board 가 guess로 바뀌게 해주는 함수
  for i in idxList:
    boardList[i] = guess
  return boardList
```

```python
# play the game 
# -------------------------------
# 1. 초기값 설정
Intro()
ans = RnWord()                    # 정답이 될 단어 랜덤선택
board = Board(ans)                # 초기 보드생성
guessHistory = ''                 # 추측 목록
manState = 0                      # 목숨
gameType = 'play'                 # 게임상태 play == 진행 / end == 종료

# -------------------------------
# 2. 무한루프 게임실행
while gameType == 'play':
  
  # -------------------------------
  # 3. 반복 출력

  # 3-1) 단두대 출력
  print(man[manState])
  # print('manState Debugging :',manState) #debugging

  # 3-2) 추측에 실패한 단어들
  print("Missing letters :",guessHistory)

  # 3-3) 찾아야 될 단어 blank ('_')로 출력
  print(board)
  print("Guess a letter.")

  # 3-4) 값 입력 받아 저장하기
  guess = input()                 # 입력받기

  #------------------------------- 
  # 4. 예측한 단어가 이전에 예측했던 적이 있는지 확인
  if guessHistory.find(guess) > 0 or board.count(guess) > 0:
    print('You have already guessed that letter. Choose again.')
    continue
  else:

    #-------------------------------
    # 5. 예측한 단어가 맞는지 확인
    if ans.find(guess)>-1:          # 맞췄을 때
      idx = Find(ans, guess)        # 맞춘 낱말 인덱스
      board = Correct(board, idx, guess)
    else:
      guessHistory = guessHistory+guess
      manState +=1



  
  
  
  #-------------------------------
  # 6. 게임 승패 결정

  # 6-1)게임오버
  if manState == 6:
    print(man[manState])
    print("Game over!")


    print("Do you want to play again? (yes or no)")
    playAgain = input()
    if playAgain == 'yes':
                                        # 게임 초기값 초기화시키기
      ans = RnWord()                    # 정답이 될 단어 랜덤선택
      board = Board(ans)                # 초기 보드생성
      guessHistory = ''                 # 추측 목록
      manState = 0                      # 목숨
    else:
      gameType = 'end'              # 종료싸인
      break
  
  # 6-2)게임 승리
  if board.count('_')==0:
    print("You win!")
    print("Do you want to play again? (yes or no)")
    playAgain = input()
    if playAgain == 'yes':
                                        # 게임 초기값 초기화시키기
      ans = RnWord()                    # 정답이 될 단어 랜덤선택
      board = Board(ans)                # 초기 보드생성
      guessHistory = ''                 # 추측 목록
      manState = 0                      # 목숨
    else:
      gameType = 'end'              # 종료싸인
      break
```
