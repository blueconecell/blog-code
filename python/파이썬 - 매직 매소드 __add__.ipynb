- Add객체 끼리는 +를 사용할 수 없다고 한다.
- +을 사용할 수 있게 해주는 매직 매소드인 ```__add__```를 Add클래스에 추가해 준다.

````python
class Add:
    def __init__(self, num):
        self.num = num
    def __add__(self, other):
        add_result = self.num + other
        return add_result
````

````python
b = Add(1)
c = Add(2)
print(b+c)
````

- 정수와 Add객체와의 덧셈이 불가능하다고 오류가 나온다.

````python
b = Add(1)
c = Add('2')
print(b+c)
````

````python
b = Add('1')
c = Add(2)
print(b+c)
````

- b에다가 c를 더하는 것이기 때문에 b는 int라고 인식하였고 c는 Add객체라고 인식하였다.


````python
b = Add(1)
# c = Add(2)
print(b+2)
````

- int라고 인식하는 b에 int를 더해주니 오류가 나오지 않았다.

````python
b = Add(1)
# c = Add(2)
print(2+b)
````

- 반면 덧셈하는 순서를 바꾸니 오류가 떴다.
- 덧셈을 진행할 때 뒤에오는 피연산자가 Add의 객체일 경우 항상 같은 오류가 발생했다.

````python
class Add:
    def __init__(self, num):
        self.num = num
    def __add__(self, other):
        add_result = self.num + other
        return add_result
    def __radd__(self, other):
        add_result = self.num + other
        return add_result
````

````python
b = Add(1)
# c = Add(2)
print(2+b)
````

- 역순으로 계산할 경우 오류가 뜨는 문제를 `__radd__`을 클래스 내에 정의하여 문제를 해결하였다.

````python
b = Add(1)
c = Add(2)
print(b+c)
````
